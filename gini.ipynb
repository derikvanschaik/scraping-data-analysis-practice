```python
import pandas as pd
import matplotlib.pyplot as plt
import os
```

```python
# data downloaded from: https://data.worldbank.org/indicator/SI.POV.GINI?end=2019&start=2019&view=map
df = pd.read_csv( os.getcwd()+'\ginidata.csv' )
```

```python
df.head()
```

```python
# We want to drop rows 0, 1, and 2 from the dataframe
df = df.drop([0, 1, 2])
df.head()
```

```python
df.columns = df.iloc[0]
df = df.iloc[1: , :] # drop the first row as it is the same as our columns row 
df.head()
```

```python
countries = pd.pivot_table(data=df,index=['Country Name'])
countries
```

```python
ginis = {} # country name -> list of gini values 
for i in range(len(countries.index)):
    series = countries.iloc[i].dropna()
    ginis[series.name] = list( series )
    
```

```python

def mean(list):
    return sum(list)/len(list)

ginis_mean = {}
for country, ginis_list in ginis.items():
    ginis_mean[country] = mean(ginis_list)

# get the top 10 highest gini mean 
top_10 = sorted(ginis_mean, key=ginis_mean.get, reverse=True)[:10]
top_10
```

```python
# plot the highest ginis to see how they have differed over the years 
#(remember the series data is chronological in df countries)
def plot(start_idx, end_idx):
    fig = plt.figure()
    ax = plt.axes()
    for country in top_10[start_idx:end_idx]:
        x = list(range(len(ginis[country])))
        y = ginis[country]
        ax.plot(x, y, label=country)
    ax.legend()
    
# split the plotting or else plots will be too crammed
plot(0, 5)
plot(5, -1) 
```

```python
# Interesting to observe that all gini indices are decreasing except for South Africa which has actually increased
# We want these gini indices to decrease -- singifies wealth inequality is decreasing in these countries 
```

```python
# I want to grab the last recorded Ginis of Each country and their respective year
ginis_latest = {} 
for i in range(len(countries.index)):
    series = countries.iloc[i].dropna()
    if len(series) > 0:
        ginis_latest[series.name] = (series.index.values[-1], series[-1]) # year, gini value 
```

```python
# from lowest latest gini to highest 
sorted_gini_latest = sorted(ginis_latest.items(), key = lambda x: x[1][1] )
highest_3 = sorted_gini_latest[-3:]
lowest_3 = sorted_gini_latest[:3]
```

```python
fig = plt.figure()
ax = plt.axes()
countries = [country_info[0] for country_info in highest_3 + lowest_3]
ginis = [country_info[1][1] for country_info in highest_3 + lowest_3]
ax.bar(countries, ginis, color = ['#FF7F7F' for i in range(3)] + ['#ADD8E6' for i in range(3)])
plt.xticks(rotation = 45) 
plt.show()
```
